```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv('3407_dublin-city_houses_df.csv')
df1.head(5)
```

```python
df1.shape
```

```python
df2 = df1.dropna()
df2.isnull().sum()
```

```python
df2.shape
```

```python
df2['price'].unique()
```

```python
df2['price'] = df2['price'].str.replace('MV: €', '')
```

```python
df2['price'].unique()
```

```python
df2['price'] = df2['price'].str.strip()
```

```python
df2['price'].unique()
```

```python
df2['price'] = df2['price'].str.split(' ').str[0]
df2['price'] = df2['price'].str.replace(',', '')
df3 = df2[df2['price'].str.contains('rice') == False]
df3['price'].astype(int)
```

```python
df3['price'].unique()
```

```python
df4 = df3.copy()
df4['beds'] = df4['beds'].str.replace(' Bed', '')
df4['baths'] = df4['baths'].str.replace(' Bath', '')
df4['beds'].astype(int)
```

```python
df4['baths'].astype(int)
```

```python
df4.shape
```

```python
df4.head(5)
```

```python
df4['floor_area'].unique()
```

```python
df5 = df4[df4['floor_area'].str.contains('ac') == False]
df5['floor_area'].astype(int)
```

```python
df5.head(15)
```

```python
df6 = df5.copy()
```

```python
df6.head(15)
```

```python
df6.reset_index(inplace=True, drop = True)
```

```python
df6.head(15)
```

```python
for i in range(df6['address'].size):
    old_location = df6['address'][i]
    string_list = str(df6['address'][i]).split(',')
    new_location = string_list[len(string_list) - 1]
    df6.loc[i,'address'] = new_location
```

```python
df6.head(5)
```

```python
df6['address'].unique()
```

```python
df6.groupby('address')['address'].agg('count')
```

```python
# Removing empty spaces from "address" variables
df6['address'] = df6['address'].str.strip()
```

```python
df6.loc[df6['address'] == 'Northern Cross', 'address'] = 'Dublin 17'
df6.loc[df6['address'] == 'Belmayne', 'address'] = 'Dublin 17'
```

```python
df6.groupby('address')['address'].agg('count')
```

```python
dummies = pd.get_dummies(df6.address)
dummies.head(3)
```

```python
df7 = pd.concat([df6, dummies], axis='columns')
df7.head(3)
```

```python
df7['Dublin 1'].describe()
```

```python
df7.floor_area = df7['floor_area'].astype(int)
```

```python
df7.floor_area.describe()
```

```python
fig = plt.figure(figsize =(10, 7))
fig1, ax1 = plt.subplots()
plt.boxplot(df7.floor_area)
plt.xlabel('Floor Area')
```

```python
df7.shape
```

```python
df8 = df7[df7['floor_area'] < 1630]
df8.shape
```

```python
fig = plt.figure(figsize =(10, 7))
fig1, ax1 = plt.subplots()
plt.boxplot(df8.floor_area)
plt.xlabel('Floor Area')
```

```python
plt.scatter(df8.floor_area, df8.price, alpha=0.5)
plt.xlabel('Floor Area')
plt.ylabel('Price')
plt.title('Price - Floor Area')
plt.show()
```

```python
plt.hist(df8.floor_area, rwidth=0.8)
plt.xlabel('Floor Area')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = df8.drop(['address','property_type' ,'price'], axis='columns')
y = df8.price
```

```python
model = LinearRegression()
model.fit(X, y)
model.score(X, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
predictions = model.predict(X)
predictions
```

```python
import pickle
with open('dublin_property_price_prediction.pickle', 'wb') as f:
    pickle.dump(model, f)

import json
columns = {
    'data_columns' : [col.lower() for col in X.columns]
}
with open('columns.json', 'w') as f:
    f.write(json.dumps(columns))
```
